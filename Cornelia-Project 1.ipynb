```python
#Dependencies
import os
import csv
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt  
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
#Pulling csv file information
dirname = "City of Atlanta unemp data"
df = []
for i in os.walk("City of Atlanta unemp data"):
    for j in i[2]:
        csvpath = os.path.join(dirname,j)
        print(csvpath)
        temp = pd.read_csv(csvpath)
        print (temp)
        df.append(temp)
```

```python
len(df)
```

```python
#compiled all of the csv files into 1 dataframe
new_df = pd.concat(df)
new_df
```

```python
#Created dataframe with only the dates by month and year
monthlyDF= new_df.loc[new_df["Timeperiod"].str.contains(",")]
monthlyDF
```

```python
# Confirming that previous dataframe pulled all the right information
meanDF=new_df.loc[~(new_df["Timeperiod"].str.contains(","))]
meanDF
```

```python
meanDF.columns
```

```python
#column variable with all of the desired columns for the new dataframe
cols = ["Periodyear","Timeperiod","Unemp","Unemprate"]
```

```python
#Dataframe with the average unemployment rate for each year
newMeanDF=meanDF[cols].set_index("Timeperiod")
newMeanDF
```

```python
#Created a new dataframe with the time period as the initial index
newMonthlyDF= monthlyDF[cols].set_index("Timeperiod")
newMonthlyDF
```

```python
#Added a column with the months changed to numbers "01,02,etc."
res = []
translate={
   "January":"01",
   "February":"02",
    "March":"03",
    "April":"04",
    "May":"05",
    "June":"06",
    "July":"07",
    "August":"08",
    "September":"09",
    "October":"10",
    "November":"11",
    "December":"12"
}

for i in newMonthlyDF.index:
    month = i.split(",")[0]
    res.append(translate [month])
newMonthlyDF["Month"]=res
newMonthlyDF
```

```python
totalyearDF = newMonthlyDF.groupby(['Periodyear'],as_index=False).sum()
totalyearDF = totalyearDF.drop(columns=['Unemprate'])

totalyearDF
```

```python
#Merging unemployment dataframe and crime data dataframe
# importing crime data csv
csvpath2= os.path.join("Crime Data.csv")
csvpath2

CrimeDF= pd.read_csv(csvpath2, low_memory=False)
```

```python
CrimeDF
```

```python
#Create dataframe showing the average of crime per year
newCrimeDF= CrimeDF.groupby(['occur_year','UC2_Literal']).count()
newCrimeDF=newCrimeDF.drop(labels=[2000,2001,2003,2004,2005,2006,2007,2008])
newCrimeDF
```

```python
#show which columns are listed in dataframe
newCrimeDF.columns
```

```python
#Drop unrelated columns
newCrimeDF=newCrimeDF.drop(columns=['rpt_date', 'occur_date',
       'occur_day', 'occur_time', 'poss_time', 'beat', 'location', 'ibr_code',
       'neighborhood', 'npu', 'lat', 'long'])

newCrimeDF
```

```python
#reset index for  new dataframe
newCrimesDF=newCrimeDF.reset_index()
newCrimesDF
```

```python
#Grouped all offenses by year
newCrimesDF=newCrimesDF.groupby(['occur_year']).sum()
newCrimesDF
```

```python
#Removed all unrelated columns except 'offense_id'
newCrimesDF=newCrimesDF['offense_id']
newCrimesDF
```

```python
#Created dataframe with only offense id showing
newCrimesDF=pd.DataFrame(newCrimesDF)
newCrimesDF
```

```python
#reset index to add occur back to columns 
newCrimesDF=newCrimesDF.reset_index()
```

```python
newCrimesDF
```

```python
#Merged yearly Unemployment totals and yearly crime offense totals into one dataframe 
mergedDF=newCrimesDF.merge(totalyearDF, left_on="occur_year",right_on="Periodyear")
mergedDF
```

```python
#Dropped 'period year' column
mergedDF=mergedDF.drop(columns=['Periodyear'])
mergedDF
```

```python
mergedDF=mergedDF.set_index("occur_year")
```

```python
#Plotted bar chart for merged dataframe
multiplot=mergedDF.plot(kind="bar")
multiplot.set_ylabel('Total Offenses vs Unemployment Totals')
multiplot.set_title('Unemployment Totals vs Total No. of Offenses per Year')
#multiplot.set_xticklabels(mergedDF["occur_year"],rotation=45)
plt.savefig('Impact of Unemployment On Crime Totals by Year.png')
plt.show()
```

```python
# Plotted a line graph to show unemployment vs number of offenses 
multiplot2=mergedDF.plot.line()
multiplot2.set_ylabel('Total Offenses vs Unemployment Totals')
multiplot2.set_title('Unemployment Totals vs Total No. of Offenses per Year')
#multiplot.set_xticklabels(mergedDF["occur_year"],rotation=45)
plt.savefig('Impact of Unemployment On Crime Totals (by Year).png')
plt.show()
```
